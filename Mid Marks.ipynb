```python
import numpy as np
import matplotlib.pyplot as plt

def predict(X, theta0, theta1):
    """Returns predictions based on current weights."""
    return theta0 + theta1 * X
```

```python
def mse(Y, Y_pred):
    """Computes Mean Squared Error."""
    m = len(Y)
    return (1 / m) * np.sum(np.square(Y_pred - Y))
```

```python
def calc_gradient(X, Y, Y_pred):
    """Calculates gradients for theta0 and theta1."""
    m = len(Y)
    d_theta0 = np.sum(Y_pred - Y) / m
    d_theta1 = np.sum((Y_pred - Y) * X) / m
    
    return d_theta0, d_theta1
```

```python
def update_weights(theta0, theta1, d_theta0, d_theta1, alpha):
    """Updates weights using gradient descent."""
    theta0 -= alpha * d_theta0
    theta1 -= alpha * d_theta1
    return theta0, theta1
```
