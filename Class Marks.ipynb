```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame creation (you should replace this with your actual DataFrame)
# df = pd.DataFrame({'Total': [5, 10, 15, 25, 35, 45]})

# Categorize total scores
df['Score_Category'] = pd.cut(df['Total'], bins=[0, 10, 20, 30, 40], labels=['0-10', '11-20', '21-30', '31-40'], right=False)

# Count categories
score_category_counts = df['Score_Category'].value_counts()

# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(score_category_counts, labels=score_category_counts.index, autopct='%1.1f%%')
plt.title("Distribution of Total Score Categories")
plt.show()
```

Categorize total scores and visualize the distribution using a pie chart

```python
# Histogram for TOTAL scores
plt.figure(figsize=(8, 5))
plt.hist(df['Total'], bins=10, color='blue', edgecolor='black')
plt.title("Distribution of Total Scores")
plt.xlabel("Total Scores")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

This plot will show the distribution of the Total column purely relying on Matplotlib

```python
questions = ["Q1aM4", "Q1bM6"]
for question in questions:
 plt.figure(figsize=(8, 5))
 plt.hist(df[question], bins=10, color='green', edgecolor='black', alpha=0.7)
 plt.title(f"Distribution of Scores for {question}")
 plt.xlabel("Scores")
 plt.ylabel("Frequency")
 plt.grid(axis='y', linestyle='--', alpha=0.7)
 plt.show()
```

```python
plt.figure(figsize=(9,8))
plt.hist(df['Q2aM6'].dropna(), bins=5, color='purple', edgecolor='black')
plt.title('Distribution of Marks in Q1aM4')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
df['Q2']=df['Q2aM6']+df['Q2bM4']
a1=df.groupby('Q2')['Total']
a1.hist(color='skyblue',figsize=[8,8],grid=False,bins=5)
plt.title("Histogram of students who scored 1-30 Marks")
plt.xlabel("Total Marks obtained")
plt.ylabel("Marks in Q2")
plt.show()
```

```python
df["Q1"] = df["Q1aM4"] + df["Q1bM6"]
df["Q2"] = df["Q2aM6"] + df["Q2bM4"]
plt.figure(figsize=(8, 6))
plt.scatter(df['Q1'], df['Q2'], color='purple')
plt.title('Scatter Plot of Marks: Q1 vs Q2')
plt.xlabel('Marks in Q1')
plt.ylabel('Marks in Q2')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
questions = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']
plt.figure(figsize=(10, 6))
for i, question in enumerate(questions, 1):
 plt.subplot(2, 3, i)
 plt.hist([question], bins=10, color='purple', edgecolor='black')
 plt.title(f'Histogram of {question}')
 plt.xlabel('Marks')
 plt.ylabel('Frequency')
 plt.grid(True)
 plt.tight_layout()
 plt.show()
```
